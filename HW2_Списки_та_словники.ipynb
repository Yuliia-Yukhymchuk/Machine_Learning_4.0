**1. Списки - базові операції**

Створіть список температур за тиждень (у градусах Цельсія) з назвою
`temperatures` і значеннями [22, 25, 19, 23, 26, 20, 21, 22].


Виконайте (кожен пункт оцінюється в 0.5 бала):
- Виведіть перший елемент
- Виведіть останній елемент
- Додайте температуру 24
- Видаліть температуру 19
- Виведіть довжину списку
- Обчисліть суму всіх температур
- Знайдіть максимальну температуру

```python
# Створення списку температур
temperatures = [22, 25, 19, 23, 26, 20, 21, 22]

# 1. Виведіть перший елемент
print("Перший елемент:", temperatures[0])

# 2. Виведіть останній елемент
print("Останній елемент:", temperatures[-1])

# 3. Додайте температуру 24
temperatures.append(24)
print("Після додавання 24:", temperatures)

# 4. Видаліть температуру 19
temperatures.remove(19)
print("Після видалення 19:", temperatures)

# 5. Виведіть довжину списку
print("Довжина списку:", len(temperatures))

# 6. Обчисліть суму всіх температур
print("Сума температур:", sum(temperatures))

# 7. Знайдіть максимальну температуру
print("Максимальна температура:", max(temperatures))
```


**2. Словники - базові операції**

Створіть словник про себе, який містить ключі `name`, `age`, `city` і значення, які відповідають вашему імені, віку і місту (можна вставити будь-які значення, не обовʼязково свої).


Виконайте (кожен пункт оцінюється в 0.5 бала):
- Виведіть значення ключа "name"
- Змініть значення ключа "age" на 21
- Додайте новий ключ "hobby" зі значенням "programming"
- Виведіть всі ключі
- Виведіть всі значення


```python
# Створення словника
person = {
    "name": "Yuliia",
    "age": 31,
    "city": "Changchun"
}

# 1. Виведіть значення ключа "name"
print("Name:", person["name"])

# 2. Змініть значення ключа "age" на 21
person["age"] = 21
print("Updated age:", person["age"])

# 3. Додайте новий ключ "hobby" зі значенням "programming"
person["hobby"] = "programming"
print("After adding hobby:", person)

# 4. Виведіть всі ключі
print("Keys:", person.keys())

# 5. Виведіть всі значення
print("Values:", person.values())
```
