### Завдання 1: Створення матриць та векторів
1. Створіть `numpy` вектор `vector` розмірністю 10 з випадковими дійсними значеннями в діапазоні від 0 до 1. Можна скористатись двома альтернативними варіантами:
  1. Генерація випадкових чисел з модулем [`random.uniform`](https://docs.python.org/3/library/random.html).
  2. Створення вектора з допомогою метода `numpy` [np.random.rand](https://numpy.org/doc/stable/reference/random/generated/numpy.random.rand.html)

2. Створіть матрицю `matrix` типу `numpy.array` розмірності 3x3 з наступними значеннями

  $$
  matrix = \begin{bmatrix}
  11 & 14 & 17 \\
  20 & 23 & 26 \\
  29 & 32 & 35
  \end{bmatrix}
  $$

  будь-яким відомими Вам способом.

Виведіть на екран вектор і матрицю, а також їх розміри.


```python
import numpy as np
import random

# 1. Вектор (через np.random.rand)
vector = np.random.rand(10)

# 2. Альтернативний варіант (через random.uniform)
vector2 = np.array([random.uniform(0, 1) for _ in range(10)])

# 3. Матриця 3x3 (задані значення)
matrix = np.array([[11, 14, 17],
                   [20, 23, 26],
                   [29, 32, 35]])

# Вивід результатів
print("Вектор (np.random.rand):")
print(vector)

print("\nВектор (random.uniform):")
print(vector2)

print("\nМатриця 3x3:")
print(matrix)

# Розміри
print("\nРозміри:")
print("vector shape:", vector.shape)
print("vector2 shape:", vector2.shape)
print("matrix shape:", matrix.shape)
```

3. Задана матриця `D` розмірністю 2х2. Подумайте спочатку, яку розмірність Ви отрмаєте, коли домножите `result_matrix` на матрицю `D` і який результат буде. А потім виконайте множення і виведіть результат на екран.

```python
A = np.array(
    [[0.18, 0.53],
     [0.75, 0.47],
     [0.35, 0.21]])

B = np.array(
    [[0.17, 0.58, 0.75],
     [0.38, 0.11, 0.15]])

C = np.array([[0], [1]])

D = np.array([[0,1], [1,0]])

display(A, B, C, D)
```

```python
# 1. Множення матриць (отримуємо 2x2)
result_matrix = np.dot(B, A)
print("Result matrix (2x2):")
print(result_matrix)


# 2. Множення на вектор (2x1)
multiply = np.dot(result_matrix, C)
print("\nResult * C (2x1):")
print(multiply)


# 3. Множення на матрицю D (2x2)
multiply2 = np.dot(result_matrix, D)
print("\nResult * D (2x2):")
print(multiply2)
```
