**1. Створення масивів**

Створіть (кожен пункт оцінюється в 0.5 бала):
- Масив зі списку `[5, 15, 25, 35, 45]`
- Масив з нулів розміром 7
- Масив з одиниць розміром 6
- Масив з числами від 10 до 19 використовуючи `np.arange()`

```python
import numpy as np

my_array = np.array([5, 15, 25, 35, 45])
zeros = np.zeros(7)  # 3x4 матриця нулів
ones = np.ones(6)    # 2x3 матриця одиниць
range_arr = np.arange(10, 19) 
print("Масив зі списку:", my_array)  
print("Масив з нулів розміром 7:", zeros) 
print("Масив з одиниць розміром 6:", ones) 
print("Масив з числами від 10 до 19 використовуючи np.arange():", range_arr) 
```

**2. Базові операції**

Створіть Numpy-масив зі значеннями [12, 24, 36, 48, 60].

Виконайте наступні операції над **оригінальним** масивом і запишіть у нову змінну результат (кожен пункт оцінюється в 0.5 бала):
- Помножте всі елементи на 2
- Додайте до всіх елементів 10
- Піднесіть всі елементи до квадрату
- Обчисліть квадратний корінь з усіх елементів

Виведіть результат виконання кожної операції.

```python
import numpy as np

my_array = np.array([12, 24, 36, 48, 60])

multiplied = my_array * 2
added = my_array + 10
squared = my_array ** 2
sqrted = my_array ** 0.5

print("Множення на 2:", multiplied)
print("Додавання 10:", added)
print("До квадрату:", squared)
print("Квадратний корінь:", sqrted)
```

**3. Статистичні функції**

Створіть Numpy-масив `data` зі значеннями [18, 27, 33, 41, 52, 68, 79].

Обчисліть використовуючи функціонал numpy (кожен пункт оцінюється в 0.5 бала):
- Середнє значення
- Мінімум
- Максимум
- Суму всіх елементів

```python
import numpy as np

data = np.array([18, 27, 33, 41, 52, 68, 79])

average = np.mean(data)
minimum = np.min(data)
maximum = np.max(data)
sum_of_all= np.sum(data)

print("Середнє значення:", average)
print("Мінімум:", minimum)
print("Максимум:", maximum)
print("Сума всіх елементів:", sum_of_all)
```

**4. Індексація**

Створіть Numpy-масив `arr` зі значеннями [15, 22, 35, 44, 58, 67].

Виведіть (кожен пункт оцінюється в 0.5 бала):
- Перший елемент
- Останній елемент
- Елементи з індексами 1 до 3
- Елементи більші за 40

```python
import numpy as np

arr = np.array([15, 22, 35, 44, 58, 67])

first_element = arr[0]
last_element = arr[-1]
elem_from_1_to_3 = arr[1:4]
bigger_than_40 = arr[arr > 40]

print("Перший елемент:", first_element)
print("Останній елемент:", last_element)
print("Елементи з індексами 1 до 3:", elem_from_1_to_3)
print("Елементи більші за 40:", bigger_than_40)
```

**5. Двовимірні масиви**

Задано матрицю 3x3.

Виконайте наступні дії з функціоналом Numpy (кожен пункт оцінюється в 0.5 бала):
- Виведіть форму (shape) матриці
- Виведіть перший рядок
- Виведіть другу колонку
- Обчисліть суму всіх елементів
- Обчисліть середнє всіх елементів

```python
matrix = np.array([[2, 4, 6],
                   [8, 10, 12],
                   [14, 16, 18]])
```

```python
print("Форма матриці:", matrix.shape)
print("Перший рядок матриці:", matrix[0])
print("Друга колонка матриці:", matrix[:, 1])
print("Сума всіх елементів матриці:", matrix.sum())
print("Середнє значення всіх елементів матриці:", matrix.mean())
```
