### Домашні завдання на відпрацювання поняття функції та її похідної

1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp

# Визначаємо символічну змінну
x = sp.symbols('x')

# Лінійна функція
f_lin = 10 * x - 5
derivative_f_lin = sp.diff(f_lin, x)
print("Похідна лінійної функції:", derivative_f_lin)
```

```python
# Визначення функції та її похідної
x = np.linspace(0, 10, 100)
f = 10 * x - 5
df = np.full_like(x, 10)  

# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x, f, label='f(x) = 10x - 5')
plt.plot(x, df, label="f'(x) = 10")

plt.xlabel('Години навчання')
plt.ylabel('Оцінка')
plt.legend()
plt.grid()
plt.show()
```

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
x = sp.symbols('x')

f_sin = sp.sin(x)
f_ln = sp.ln(x)


derivative_f_sin = sp.diff(f_sin, x)
derivative_f_ln = sp.diff(f_ln, x)


print("Похідна функції синуса:", derivative_f_sin)
print("Похідна логарифмічної функції:", derivative_f_ln)
```

```python
x = np.linspace(0.1, 10, 100)

# Функції
f_sin = np.sin(x)
df_sin = np.cos(x)

f_ln = np.log(x)
df_ln = 1 / x

# Мультиплот
plt.figure(figsize=(12, 5))

# --- sin ---
plt.subplot(1, 2, 1)
plt.plot(x, f_sin, label='sin(x)')
plt.plot(x, df_sin, label='cos(x)')
plt.title('Синус та його похідна')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()

# --- ln ---
plt.subplot(1, 2, 2)
plt.plot(x, f_ln, label='ln(x)')
plt.plot(x, df_ln, label='1/x')
plt.title('Логарифм та його похідна')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.symbols('x')
f_x = x*x - 5*x + 1
derivative_f = sp.diff(f_x, x)
print("Похідна функції:", derivative_f)
```

```python
x_vals = np.linspace(0, 100, 200)

f = x_vals**2 - 5*x_vals + 1
df = 2*x_vals - 5  

# --- Графік ---
plt.figure(figsize=(10, 5))
plt.plot(x_vals, f, label='f(x) = x² - 5x + 1')
plt.plot(x_vals, df, label="f'(x) = 2x - 5")

plt.xlabel('Кількість вироблених одиниць')
plt.ylabel('Загальні витрати')
plt.title('Функція витрат та її похідна')
plt.legend()
plt.grid()

plt.show()
```
