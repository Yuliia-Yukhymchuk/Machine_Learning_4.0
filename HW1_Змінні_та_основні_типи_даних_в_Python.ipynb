**2. Арифметичні операції**

Створіть дві змінні: `var1` зі значенням 15, `var2` зі значенням 4.

Обчисліть та виведіть (кожен пункт оцінюється в 0.5 бала):
- Суму
- Різницю
- Добуток
- Ділення
- Цілочисельне ділення
- Залишок від ділення
- Піднесення до степеня


```python
var1 = 15
var2 = 4

# Сума
print("Сума:", var1 + var2)

# Різниця
print("Різниця:", var1 - var2)

# Добуток
print("Добуток:", var1 * var2)

# Ділення
print("Ділення:", var1 / var2)

# Цілочисельне ділення
print("Цілочисельне ділення:", var1 // var2)

# Залишок від ділення
print("Залишок від ділення:", var1 % var2)

# Піднесення до степеня
print("Піднесення до степеня:", var1 ** var2)
```

**3. Робота з рядками**

Створіть зміну `last_name`, яка містять ваше прізвище.

Виконайте (кожен пункт оцінюється в 0.5 бала):
- Об'єднайте значення змінних `name` та `last_name` через пробіл в один рядок і збережіть в нову змінну `full_name`
- Виведіть `full_name` великими літерами
- Виведіть `full_name` малими літерами
- Виведіть довжину `full_name`
- Виведіть фразу: "Мене звати [ім'я], мені [вік] років". Вік використайте зі змінної, створеної у вправі 1.


```python
name = "Yuliia"      # ім'я зі вправи 1
age = 31             # вік зі змінної у вправі 1
last_name = "Yukhymchuk"   

# Об'єднання name і last_name
full_name = name + " " + last_name

# Вивід full_name великими літерами
print(full_name.upper())

# Вивід full_name малими літерами
print(full_name.lower())

# Довжина full_name
print("Довжина full_name:", len(full_name))

# Вивід фрази з ім'ям і віком
print(f"Мене звати {name}, мені {age} років")
```
