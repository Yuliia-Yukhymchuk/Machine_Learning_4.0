**Про датасет:**

Tips - це датасет, який містить інформацію про рахунки та чайові в ресторані.

**Колонки датасету:**
- `total_bill` - загальна сума рахунку (долари)

**Джерело:** Оригінальні дані зібрані в одному з американських ресторанів у 1990-х роках. Скачати і переглянути набір даних можна тут: https://www.kaggle.com/datasets/sakshisatre/tips-dataset

**1. Завантаження та огляд**

Завантажте датасет Tips (Чайові) з файла у матеріалах `tips.csv`.

- Перші 5 рядків
- Розмір датасету (shape)

```python
import pandas as pd

# Завантаження датасету Tips
df = pd.read_csv('tips.csv')

print("Перші 5 рядків датасету:")
display(df.head(5))

print("\nРозмір датасету (rows, columns):")
print(df.shape)

print("\nНазви колонок:")
display(df.columns)

print("\nІнформація про типи даних:")
df.info()
```

Виберіть та виведіть (кожен пункт оцінюється в 0.5 бала):
- Колонку `total_bill` (загальний рахунок)

```python
print("Колонка total_bill:")
display(df['total_bill'])

print("\nКолонки total_bill та tip:")
display(df[['total_bill', 'tip']])

print("\nПерші 10 рядків датасету:")
display(df.iloc[:10])

print("\nРядок з індексом 5:")
display(df.iloc[5])
```

**3. Фільтрація даних**

Відфільтруйте повний набір даних та виведіть (кожен пункт оцінюється в 0.5 бала):
- Всі рахунки більше 20 доларів (`total_bill > 20`)
- Всі відвідувачі-жінки (`sex == 'Female'`)
- Рахунки на вечерю (`time == 'Dinner'`)
- Рахунки на вечерю з сумою чеку більше 30 доларів

Для виведення результатів фільтрації:
- зберегти результат у окрему змінну
- показати кількість записів у відфільтрованому наборі
- вивести перші 5 записів відфільтрованого набору.

```python
# 1. Рахунки більше 20 доларів
print("Рахунки більше 20 доларів:")
bills_more_20 = df[df['total_bill'] > 20]
print("Кількість записів:", bills_more_20.shape[0])
display(bills_more_20.head(5))


# 2. Всі відвідувачі-жінки
print("\nВсі відвідувачі-жінки:")
female_guests = df[df['sex'] == 'Female']
print("Кількість записів:", female_guests.shape[0])
display(female_guests.head(5))


# 3. Рахунки на вечерю
print("\nРахунки на вечерю:")
dinner_bills = df[df['time'] == 'Dinner']
print("Кількість записів:", dinner_bills.shape[0])
display(dinner_bills.head(5))


# 4. Рахунки на вечерю з сумою більше 30 доларів
print("\nРахунки на вечерю з сумою більше 30 доларів:")
dinner_bills_more_30 = df[
    (df['time'] == 'Dinner') & (df['total_bill'] > 30)
]
print("Кількість записів:", dinner_bills_more_30.shape[0])
display(dinner_bills_more_30.head(5))
```

**4. Описова статистика**

Обчисліть (кожен пункт оцінюється в 0.5 бала):
- Середній розмір рахунку на всьому наборі даних
- Максимальні чайові
- Кількість пропущених значень у всіх колонках
- Скільки в наборі даних рахунків за обіди і за вечері

```python
# 1. Середній розмір рахунку
print("Середній розмір рахунку:")
average_bill = df['total_bill'].mean()
print(average_bill)


# 2. Максимальні чайові
print("\nМаксимальні чайові:")
max_tip = df['tip'].max()
print(max_tip)


# 3. Кількість пропущених значень у всіх колонках
print("\nКількість пропущених значень у кожній колонці:")
missing_values = df.isnull().sum()
display(missing_values)


# 4. Кількість рахунків за обіди та вечері
print("\nКількість рахунків за обіди та вечері:")
lunch_dinner_counts = df['time'].value_counts()
display(lunch_dinner_counts)
```

---

**5. Статистика із групуванням**

Обчисліть (кожен пункт оцінюється в 0.5 бала):
- Середній рахунок за статтю (groupby `sex`)
- Середні чайові за днем тижня (groupby `day`)
- Скільки відвідувачів у кожен день

```python
# 1. Середній рахунок за статтю
average_bill_by_sex = df.groupby('sex')['total_bill'].mean()
print("Середній рахунок за статтю:")
display(average_bill_by_sex)


# 2. Середні чайові за днем тижня
average_tip_by_day = df.groupby('day')['tip'].mean()
print("\nСередні чайові за днем тижня:")
display(average_tip_by_day)


# 3. Кількість відвідувачів у кожен день
guests_by_day = df.groupby('day')['sex'].count()
print("\nКількість відвідувачів у кожен день:")
display(guests_by_day)
```

**6. Створення нової колонки**

Створіть нову колонку `tip_percentage` - відсоток чайових від загальної суми рахунку.
Виведіть перші 10 рядків з новою колонкою.

```python
# Створення нової колонки tip_percentage
df['tip_percentage'] = (df['tip'] / df['total_bill'] * 100).round(1)

print("Перші 10 рядків з новою колонкою tip_percentage:")
display(df.head(10))
```
